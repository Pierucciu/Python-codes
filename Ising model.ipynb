```python
from numpy.random import randint,seed
from numpy import zeros,arange,sum
from math import exp
from random import random
import matplotlib.pyplot as plt

#constant
N=20
J=1
T=1
k=1
t_max=1000000
energies=[]
magnetization=[]

a = randint(2,size=(N, N))    
a[a==0] = -1
b = zeros((N+2, N+2))
b[1:-1, 1:-1] = a

def energy(b):
    tot=0
    for i in range(1,N+1):
        for j in range(1,N+1):
            tot=tot+b[i,j]*(b[i+1,j]+b[i,j+1])
        
    total_energy=-J*tot
    return total_energy

E=energy(b)
energies.append(E)
magnetization.append(sum(b))
```

```python
t=0
while t<t_max:
    i=randint(1,N+1)
    j=randint(1,N+1)
    dE=J*2*b[i,j]*(b[i+1,j]+b[i,j+1]+b[i-1,j]+b[i,j-1])
    t+=1

    if random()<exp(-dE/(k*T)):
        b[i,j]=-b[i,j]
        E=E+dE
        energies.append(E)
        magnetization.append(sum(b))
                            
times=arange(len(energies))
```

```python
plt.plot(times,energies)
```

```python
plt.plot(times,magnetization)
```
